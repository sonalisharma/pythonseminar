## Getting Frequency to notes mapping table by scrapping website

```python
import numpy
import BeautifulSoup
import urllib2
freq_notes={}   
url = "http://www.phy.mtu.edu/~suits/notefreqs.html"
data = urllib2.urlopen(url).read()
soup = BeautifulSoup.BeautifulSoup(data)
rows = soup.findAll('tr')
data = [[td.findChildren(text=True) for td in tr.findAll("td")] for tr in rows]
for d in data:
    if(d):
        freq_notes[float(str("".join(d[1])).strip())]=str("".join(d[0]))

freq_notes    
```

```python
# Frequency Graph

%pylab inline
from scipy import *
from scipy import fftpack

def maxfrequency(f,p):
    maxvalue = f[p.argmax()]
    diff1 = 1000
    for k in freq_notes.keys():
        if abs(k - maxvalue) < diff1:
            key =k
            diff1 = abs(k-maxvalue)
    print freq_notes[key] , key
    
def plots(f,p):
    plot_title = ("Frequency Audio Analysis")
    fig = plt.figure(figsize=(7, 4.5))
    ax1 = fig.add_subplot(1,1,1)
    ax1.plot(f,p, color="blue", linestyle="-")
    ax1.set_xlabel("Frequency [Hz]")
    ax1.set_ylabel("Power Spectrum")
    ax1.set_xlim(0, 2000)
    ax1.set_title(plot_title)
    show()
    
#Iterating through files   
soundfiles = listdir("sound_files")
for name in soundfiles:
    if name!=".DS_Store":
        print "sound_files/"+name
        f = aifc.open("sound_files/"+name)
        data = f.readframes(f.getnframes())
        integer_data = fromstring(data, dtype=np.int32)
        RATE =  f.getframerate()
        time = arange(size(integer_data)) / float(RATE)        
        sig_fft = np.abs(np.fft.fft(integer_data))**2       
        samplefreq = np.fft.fftfreq(integer_data.size,1/float(RATE))
        pidxs = np.where(samplefreq > 0)
        freqs = samplefreq[pidxs]
        power = np.abs(sig_fft)[pidxs] 
        maxfrequency(freqs,power)
        plots(freqs,power)
```
